```python
import cv2
from google.colab.patches import cv2_imshow
def display_image(img):
  """
  As colab doesn't support cv2.imshow they have made a patch for it and hence we need to use that to 
  display the image
  """
  cv2_imshow(img)
```

```python
img=cv2.imread("/content/butterfly.jpg")
display_image(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display_image(gray)
```

```python
inverted = 255 - gray
display_image(inverted)
```

```python
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
display_image(blurred)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray, inverted_blurred, scale=256)
display_image(pencil_sketch)
```
